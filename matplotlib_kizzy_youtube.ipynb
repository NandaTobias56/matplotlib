```python
import numpy as np
x = np.arange(0, 10,0.1)
fig1, fl_axes = plt.subplots (ncols=2, nrows=2)
fl_axes[0,0].plot(np.sin(x)) #funo seno
fl_axes[0,1].plot(np.cos(x)) #funo coseno
fl_axes[1,0].plot(np.tan(x)) #funo tangente
```

```python
x = np.arange(0, 10,0.1)
fig1, fl_axes = plt.subplots (ncols=2, nrows=2)
fl_axes[0,0].plot(np.sin(x)) #funo seno
fl_axes[0,0].plot(np.cos(x)) #funo cosseno plotando no mesmo grfico
fl_axes[0,1].plot(np.cos(x)) #funo cosseno
fl_axes[1,0].plot(np.tan(x)) #funo tangente
```

```python
x = np.arange(0, 10,0.1)
fig1, fl_axes = plt.subplots (ncols=2, nrows=2, figsize=(15,10)) #aumentando o tamanho da figura com figsize
fig1.suptitle("Minha primeira figura", size=20) #dando ttulo para a figura e aumentando seu tamanho

fl_axes[0,0].plot(np.sin(x)) #funo seno
fl_axes[0,0].plot(np.cos(x)) #funo cosseno plotando no mesmo grfico
fl_axes[0,0].set_title("Caixa 1") # dando ttulo para o grfico
fl_axes[0,0].set_xlabel("Caixa 1 - Label X") # dando nome aos eixos
fl_axes[0,0].set_ylabel("Caixa 1 - Label Y")

fl_axes[0,1].plot(np.cos(x)) #funo cosseno
fl_axes[0,1].set_title("Caixa 2") # dando ttulo para o grfico

fl_axes[1,0].plot(np.tan(x)) #funo tangente
fl_axes[1,0].set_title("Caixa 3") # dando ttulo para o grfico

fl_axes[1,1].set_title("Caixa 4") # dando ttulo para o grfico
```

```python
x = np.arange(0, 10,0.1)
fig1, fl_axes = plt.subplots (ncols=2, nrows=2, figsize=(15,10)) #aumentando o tamanho da figura com figsize
fig1.suptitle("Minha primeira figura", size=20) #dando ttulo para a figura e aumentando seu tamanho


fl_axes[0,0].plot(np.sin(x), label="sen(x)", color="hotpink") #adicionando legenda com o label e trocando cores
fl_axes[0,0].plot(np.cos(x), label="cos(x)", color="blue")
fl_axes[0,0].set_title("Caixa 1") # dando ttulo para o grfico
fl_axes[0,0].set_xlabel("Caixa 1 - Label X") # dando nome aos eixos
fl_axes[0,0].set_ylabel("Caixa 1 - Label Y")
fl_axes[0,0].legend()#chamando a funo para legenda no grfico


fl_axes[0,1].plot(np.cos(x)) #funo cosseno
fl_axes[0,1].set_title("Caixa 2") # dando ttulo para o grfico

fl_axes[1,0].plot(np.tan(x)) #funo tangente
fl_axes[1,0].set_title("Caixa 3") # dando ttulo para o grfico

fl_axes[1,1].set_title("Caixa 4") # dando ttulo para o grfico
```
